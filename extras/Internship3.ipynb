```python
l=[1,2,3,4,5,6,7]
for i in range(len(l)+1):
    print (l[i])
```

```python
l=[1,2,3,4,5,6,7]
for i in range(len(l)+1):
    try:
        print (l[i])
    except IndexError:
        print("Index Exceeded")
        break
```

```python
#if you don't know what is the error. e is error
l=[1,2,3,4,5,6,7]
for i in range(len(l)+1):
    try:
        print (l[i])
    except Exception as e:
        print(e)
```

```python
# Cells that include finally will always be executed
l=[1,2,3,4,5,6,7]
for i in range(len(l)+1):
    try:
        print (l[i])
    except Exception as e:
        print(e)
    finally:
        print("Execution successful")
```
