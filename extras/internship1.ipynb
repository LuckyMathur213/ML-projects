```python
my_guess=10
your_guess=int(input("your guess: "))
while your_guess != my_guess:
     if your_guess > my_guess:
        print ("Lower")
     elif your_guess < my_guess:
        print("Higher")
     your_guess=int(input("Next guess: "))

print("You got it")
```
