```python
def compare_to_five(y):
    if y > 5:
        return "Greater"
    elif y < 0:
        return "Negative"
    elif y < 5:
        return "Less"
    else:
        return "Equal"
```

```python
print (compare_to_five(10))
```

```python
print (compare_to_five(2))
```

```python
print (compare_to_five(5))
```

```python
print (compare_to_five(-3))
```
