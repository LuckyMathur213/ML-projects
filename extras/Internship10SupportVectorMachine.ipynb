```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()

#Apply a fix to the statsmodels library
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
```

```python
data
```

```python
x=data['EstimatedSalary']
y=data['Purchased']
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#data could be scaled but it's not necessary
from sklearn.preprocessing import StandardScaler
```

```python
x_train = x_train.values.reshape(-1, 1) if isinstance(x_train, pd.Series) else x_train
x_test = x_test.values.reshape(-1, 1) if isinstance(x_test, pd.Series) else x_test

sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0) 
```

```python
classifier.fit(x_train, y_train)
```

```python
#scale values back to original form
y_pred=classifier.predict(x_test)
```

```python
y_pred = y_pred.to_numpy().reshape(len(y_pred), 1) if isinstance(y_pred, pd.Series) else y_pred.reshape(len(y_pred), 1)
y_test = y_test.to_numpy().reshape(len(y_test), 1) if isinstance(y_test, pd.Series) else y_test.reshape(len(y_test), 1)

# Concatenate the reshaped arrays
result = np.concatenate((y_pred, y_test), axis=1)
print(result)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm=confusion_matrix(y_test, y_pred)
```
