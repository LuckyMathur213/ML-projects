```python
# * is args. A special type of f is args f
def add_many(*w):
    s=0
    for i in w:
        s=s+i
    return s
```

```python
add_many(3,66,2,5,6,8)
```

```python
# Self parameter can be used anywhere inside the class. It refers to that instance of the class for which it has been initialised, ex A, A is an object for the class car
class Car:
    def __init__(self, car_name, car_color):
        self.namecar=car_name
        self.colorcar=car_color

    def engine(self, eng="900 horsepower"):
        print(f"The car {self.namecar} who's color is {self.colorcar} has an engine of {eng}")
```

```python
A=Car("Tata","Green")
```

```python
A.engine()
```

```python
#to inherit means to use a class's method in a diff class. Class within a class is a child class, the OG class is super class. Super allows use of parent vars
#Child class
class Innova(Car):
    def __init__(self, music_name, car_name, car_color):
        super().__init__(car_name, car_color)
        self.musicsystem=music_name

    def music_system(self):
        print(f"Innova has an awesome music system name {self.musicsystem} with car color {self.colorcar}")
```

```python
obj=Innova("Dolby","Innova Luxury","Black")
```

```python
obj.music_system()
```

```python
obj.engine()
```
