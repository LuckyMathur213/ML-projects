```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('Mall_customers.csv')
data
```

```python
x=data.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,30):
    km=KMeans(n_clusters=i, random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,30),wcss)
plt.xticks(np.arange(1,29,step=1))
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
km=KMeans(5, random_state=42)
```

```python
y_means=km.fit_predict(x)
```

```python
plt.scatter(x[y_means==0,0],x[y_means==0,1],s=100,c='red',label='clu 1')
plt.scatter(x[y_means==1,0],x[y_means==1,1],s=100,c='blue',label='clu 2')
plt.scatter(x[y_means==2,0],x[y_means==2,1],s=100,c='green',label='clu 3')
plt.scatter(x[y_means==3,0],x[y_means==3,1],s=100,c='cyan',label='clu 4')
plt.scatter(x[y_means==4,0],x[y_means==4,1],s=100,c='magenta',label='clu 5')
```
