```python
def wage(hours):
    return hours*25

def with_bonus(hours):
    return wage(hours) + 50
```

```python
def add_10(m):
    if m >= 100:
        m = m + 10
        return m
    else:
        return "Save more"
```

```python
def compare_the_two(x, y):
    if x > y:
        print("Greater")
    if y > x:
        print("Less")
    if x == y:
        print("Equal")
```

```python
compare_the_two(1, 3)
```

```python
def subtract_bc(a, b, c):
    result = a - b * c
    print ("Parameter a =", a)
    print ("Parameter b =", b)
    print ("Parameter c =", c)
    return result
```

```python
subtract_bc(10 , 3, 2)
```

```python
subtract_bc(b=5 , a=7, c=9)
```

```python
def distance_from_zero(x):
    if type(x)==int:
        return abs(x)
    else:
        print("Not possible")
```
