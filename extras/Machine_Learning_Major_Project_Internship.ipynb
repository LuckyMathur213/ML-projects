```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

```python
data=pd.read_csv('fraud_detection1.csv')
```

```python
data
```

```python
data1=data.drop(['Unnamed: 0','step'],axis=1)
data1.describe(include='all')
```

```python
y=data1['isFraud']
x=data1[['type','oldbalanceorg',	'newbalanceOrig','oldbalanceDest','newbalanceDest']]	
```

```python
x_dum= pd.get_dummies(x, drop_first=True)
```

```python
x_dum
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x_dum,y, test_size=0.6, random_state=0)
x_train = x_train.astype({'type_PAYMENT': int, 'type_TRANSFER': int})
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm=confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
accuracy_score(y_test, y_pred)
```
