```python
def count(numbers):
    total = 0
    for x in numbers:
        if x < 20:
            total += 1
    return total
```

```python
nums = [1,35,12,24,31,51,70,100]
x=0
def count(nums):
    numbers = sorted(nums)
    tot = 0
    
    while numbers[tot] < 20:
        tot += 1
    return tot
```
